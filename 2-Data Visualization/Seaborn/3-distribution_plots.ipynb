# Distribution Plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

## Dist Plot

```python
tips = pd.read_csv('../../Datasets/tips.csv')
```

```python
sb.distplot(tips['total_bill'])
plt.show()
```

```python
sb.distplot(tips['total_bill'], bins=30)
plt.show()
```

```python
sb.distplot(tips['total_bill'], bins=30, kde=False)
plt.show()
```

```python
sb.distplot(tips['total_bill'], hist=False)  # line plot
plt.show()
```

## Joint Plot

```python
tips = pd.read_csv('../../Datasets/tips.csv')
```

```python
sb.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
plt.show()
```

```python
sb.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
plt.show()
```

```python
sb.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
plt.show()
```

## Pair Plot

```python
tips = pd.read_csv('../../Datasets/tips.csv')
```

```python
sb.pairplot(tips)
plt.show()
```

```python
sb.pairplot(tips, hue='sex', palette='coolwarm', diag_kind='kde', kind='scatter')
plt.show()
```

## Facet-Grid Plot

```python
iris = pd.read_csv('../../Datasets/iris.csv')
```

```python
sb.FacetGrid(iris, hue='variety', height=5).map(plt.scatter, 'sepal.length', 'sepal.width').add_legend()
plt.show()
```

```python
sb.rugplot(tips['total_bill'])
plt.show()
```
