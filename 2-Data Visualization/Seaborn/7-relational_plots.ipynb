# Relational Plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

## Rel Plot

```python
tips = sb.load_dataset('tips')
```

```python
sb.relplot(x='total_bill', y='tip', hue='day', data=tips)
plt.show()
```

```python
sb.relplot(x='total_bill', y='tip', hue='day', col='time', data=tips)
plt.show()
```

```python
sb.relplot(x='total_bill', y='tip', hue='day', col='time', row='sex', data=tips)
plt.show()
```

```python
sb.relplot(x='total_bill', y='tip', hue='day', col='time', palette=['blue', 'red'], size=(10, 100), data=tips)
plt.show()
```

```python
sb.relplot(x='total_bill', y='tip', hue='day', style='event', kind='line', height=5, data=tips)
plt.show()
```

## Scatter Plot

```python
sb.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
sb.scatterplot(x='total_bill', y='tip', hue='time', data=tips)
plt.show()
```

```python
sb.scatterplot(x='total_bill', y='tip', hue='time', marker='+', data=tips)
plt.show()
```

```python
sb.scatterplot(x='total_bill', y='tip', hue='day', style='time', palette='Set2', data=tips)
plt.show()
```

## Line Plot

```python
tips = sb.load_dataset('tips')
```

```python
sb.lineplot(x='total_bill', y='time', data=tips)
plt.show()
```

```python
sb.lineplot(x='total_bill', y='sex', lw=2, color='red', label='bill payed', data=tips)
plt.show()
```

```python
sb.lineplot(x='total_bill', y='sex', hue='smoker', data=tips)
plt.show()
```

```python
sb.lineplot(x='total_bill', y='sex', hue='smoker', style='event', markers=True, dashes=False, data=tips)
plt.show()
```
