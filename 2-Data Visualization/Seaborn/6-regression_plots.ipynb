# Regression Plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

## LM Plot

```python
tips = sb.load_dataset('tips')
```

```python
sb.lmplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
sb.lmplot(x='total_bill', y='tip', hue='smoker', palette='Set1', data=tips)
plt.show()
```

```python
sb.lmplot(x='total_bill', y='tip', hue='smoker', markers=['o', 'x'], palette='Set2', data=tips)
plt.show()
```

```python
sb.lmplot(
    x='total_bill',
    y='tip',
    hue='day',
    markers=['o', 'x'],
    palette={'Yes': 'green', 'No': 'blue'},
    data=tips,
    row='sex',
    col='day',
    height=7,
    x_jitter=0.1,
    col_wrap=2
)
plt.show()
```

## Reg Plot

```python
sb.regplot(x='total_bill', y='tip', data=tips)  # linear regression
plt.show()
```

```python
sb.regplot(x='total_bill', y='tip', data=tips, x_jitter=0.1)
plt.show()
```

```python
sb.regplot(x='size', y='total_bill', data=tips, x_estimator=np.mean)
plt.show()
```

```python
sb.regplot(x='size', y='total_bill', data=tips, x_estimator=np.mean, x_bins=4)
plt.show()
```

```python
sb.regplot(
    x='size',
    y='total_bill',
    data=tips,
    x_estimator=np.mean,
    logistic=True
)  # logistic regression
plt.show()
```

```python
sb.regplot(x='size', y='total_bill', data=tips, x_estimator=np.mean, order=2)  # polynomial regression
plt.show()
```
