## Merge Images
- To merge images size of images should be same

```python
image1 = Image.open('images/cv-image3.jpeg')
image1.size
```

```python
image2 = Image.open('images/cv-image4.jpeg')
image2.size
```

```python
resized_image1 = image1.resize((300, 300))
resized_image2 = image2.resize((300, 300))
merged_image_width = 2 * resized_image2.size[0]  # no of images * width of any image i.e 2 * 200 = 400
merged_image_height = resized_image1.size[1]  # height of any image as all images are of equal size
new_image = Image.new('RGB', (merged_image_width, merged_image_height))
new_image.paste(resized_image1, (0,0))  # paste at width = 0 and height = 0 (top - left)
new_image.paste(resized_image2, (300,0))  # paste at width = 200 and height = 0 (top - right)
new_image
```

## Blur an Image

```python
image = Image.open('images/arslan.jpeg')
image
```

### Normal Blur

```python
normal_blur_image = image.filter(ImageFilter.BLUR)
normal_blur_image
```

### Box Blur
- syntax = ImageFilter.BoxBlur(radius)

```python
normal_blur_image = image.filter(ImageFilter.BoxBlur(5))
normal_blur_image
```

```python
normal_blur_image = image.filter(ImageFilter.BoxBlur(10))
normal_blur_image
```

```python
normal_blur_image = image.filter(ImageFilter.BoxBlur(15))
normal_blur_image
```
