# Grid Plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

## Pair Grid

```python
iris = sb.load_dataset('iris')
```

```python
pg = sb.PairGrid(iris)
```

```python
plt.figure(figsize=(200, 200))
pg.map(plt.scatter)
plt.show()
```

```python
pg.map_diag(plt.hist)
pg.map_lower(sb.kdeplot)
pg.map_upper(plt.scatter)
plt.show()
```

## Pair Plot

```python
iris = sb.load_dataset('iris')
```

```python
sb.pairplot(iris)
plt.show()
```

```python
sb.pairplot(iris, hue='species', palette='rainbow')
plt.show()
```

## Facet Grid

```python
tips = sb.load_dataset('tips')
```

```python
fg1 = sb.FacetGrid(tips, col='time', row='smoker')
fg1.map(plt.hist, 'total_bill')
plt.show()
```

```python
fg2 = sb.FacetGrid(tips, col='time', row='smoker', hue='sex')
fg2.map(plt.scatter, 'total_bill', 'tip').add_legend()
plt.show()
```

## Joint Grid

```python
tips = sb.load_dataset('tips')
```

```python
jg = sb.JointGrid(x='total_bill', y='tip', data=tips)
jg.plot(sb.regplot, sb.distplot)
plt.show()
```
