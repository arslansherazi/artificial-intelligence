# Categorical Plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

## Bar Plot

```python
tips = pd.read_csv('../../Datasets/tips.csv')
```

#### bar plot (default estimator = mean)

```python
sb.barplot(x='sex', y='total_bill', data=tips, hue='sex')
plt.title('Mean bill payed by men and women')
plt.show()
```

```python
sb.barplot(x='sex', y='total_bill', data=tips, estimator=np.std, hue='sex')
plt.title('Standard deviation of bill payed by men and women')
plt.show()
```

## Count Plot

```python
tips = pd.read_csv('../../Datasets/tips.csv')
```

```python
sb.countplot(x='sex', data=tips, hue='sex')  # counts the occurrences of x
plt.title('Count of men and women')
plt.show()
```

## Box Plot

```python
tips = pd.read_csv('../../Datasets/tips.csv')
```

```python
sb.boxplot(x='day', y='total_bill', data=tips, palette='rainbow')
plt.title('Bill payed on different days')
plt.show()
```

```python
sb.boxplot(data=tips, palette='rainbow', orient='h')
plt.title('Distribution of attributes')
plt.show()
```

```python
sb.boxplot(x='total_bill', y='smoker', data=tips, hue='smoker', palette='coolwarm')
plt.title('Bill payed by smokers and non-smokers')
plt.show()
```

## Violin Plot

```python
tips = pd.read_csv('../../Datasets/tips.csv')
```

```python
sb.violinplot(x='day', y='total_bill', data=tips, palette='rainbow')
plt.title('Bill payed on different days')
plt.show()
```

```python
sb.violinplot(x='day', y='total_bill', hue='sex', data=tips, palette='Set1')
plt.title('Bill payed on different days')
plt.show()
```

```python
sb.violinplot(x='day', y='total_bill', hue='sex', data=tips, split=True, palette='Set2')
plt.title('Bill payed on different days')
plt.show()
```

## Cat Plot

```python
tips = pd.read_csv('../../Datasets/tips.csv')
```

### Factor Plot

```python
sb.catplot(x='sex', y='total_bill', data=tips, kind='bar')
plt.title('Bill payed by men and women')
plt.show()
```

```python
sb.catplot(x='time', y='total_bill', data=tips, kind='violin', hue='sex')
plt.title('Bill payed by men and women')
plt.show()
```

## Strip Plot

```python
tips = pd.read_csv('../../Datasets/tips.csv')
```

```python
sb.stripplot(x='day', y='total_bill', data=tips)
plt.title('Bill payed on different days')
plt.show()
```

```python
sb.stripplot(x='day', y='total_bill', data=tips, jitter=True)  # jitter adds some noise into data
plt.title('Bill payed on different days')
plt.show()
```

```python
sb.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', palette='Set1')
plt.title('Bill payed on different days')
plt.show()
```

```python
sb.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', palette='Set1', dodge=True)
plt.title('Bill payed on different days')
plt.show()
```

## Swarm Plot

```python
tips = pd.read_csv('../../Datasets/tips.csv')
```

#### swarm plot (alternative of jitter from strip plot)

```python
sb.swarmplot(x='day', y='total_bill', data=tips)
plt.title('Bill payed on different days')
plt.show()
```

```python
sb.swarmplot(x='day', y='total_bill', data=tips, hue='sex', palette='Set1')
plt.title('Bill payed on different days')
plt.show()
```

```python
sb.swarmplot(x='day', y='total_bill', data=tips, hue='sex', palette='Set1', dodge=True)
plt.title('Bill payed on different days')
plt.show()
```
