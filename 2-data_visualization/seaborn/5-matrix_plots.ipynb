# Matrix Plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

## Heat Map

```python
flights = sb.load_dataset('flights')
pv_flights = flights.pivot_table(values='passengers', index='month', columns='year')
```

```python
sb.heatmap(pv_flights, cmap='magma', linecolor='white', linewidths=1)
plt.title('No of passengers travel in flights')
plt.xlabel('years')
plt.ylabel('months')
plt.show()
```

## Cluster Map

```python
flights = sb.load_dataset('flights')
pv_flights = flights.pivot_table(values='passengers', index='month', columns='year')
```

```python
sb.clustermap(pv_flights, cmap='coolwarm', standard_scale=1)
plt.title('No of passengers travel in flights')
plt.xlabel('years')
plt.ylabel('months')
plt.show()
```
